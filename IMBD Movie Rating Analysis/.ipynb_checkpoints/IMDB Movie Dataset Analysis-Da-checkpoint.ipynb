### Load all the relevant files

```python
# read movies.csv file
movies = pd.read_csv('./ml-25m/movies.csv', sep=',')
print(type(movies))
movies.head()
```

```python
# read ratings.csv file
ratings = pd.read_csv('./ml-25m/ratings.csv', sep=',')
print(type(movies))
ratings.head()
```

### Data Prep

```python
# check if any of the timestamp is null
ratings['timestamp'].isnull().any()
```

```python
# convert unix time to datetime64
ratings['parsed_time'] = pd.to_datetime(ratings['timestamp'],unit='s')
ratings.head()
```

```python
# extract the year of ratings
ratings['year_of_rating'] = pd.DatetimeIndex(ratings['parsed_time']).year
ratings.head()
```

```python
# get the time range of the ratings - 1995 to 2019
ratings['year_of_rating'].describe()
```

```python
# check if any of the year has no enough rating records that cannot be counted as a good sample
ratings_count = ratings[['year_of_rating','rating']].groupby('year_of_rating').count()
print(ratings_count)
# below results indicate that only 3 ratings were made in 1995, so 1995 will be excluded from the analysis
```

```python
# see how the ratings are distributed
ratings['rating'].describe()
```

```python
# check if there's any null rating
ratings['rating'].isnull().any()
```

```python
# check if any null value in movie dataset
movies.isnull().any()
```

```python
# extract movie release year from title column in movies dataset
movies['year_of_release']=movies['title'].str.extract(r'\(([0-9]{4})\)')
movies.head()
```

```python
# split genres by using | as delimiter
movie_genres_df = movies['genres'].str.split('|',expand=True)
movie_genres_df.head(10)
```

```python
# split genres by using | as delimiter
# get all movie genres into a set (use union funcation to remove duplicates)
movie_genres = set()
for i in movies['genres'].to_list():
    genres_split = set(i.split('|'))
    movie_genres = movie_genres.union(genres_split)

# convert the set to list
movie_genres = list(movie_genres)
print(movie_genres)
```

```python
# remove perenthesis from movie_genres list
for i in movie_genres:
    movie_genres[movie_genres.index(i)] = i.replace('(','').replace(')','')
print(movie_genres)
```

### Data Analysis

```python
# join movie dataset with rating dataset
movie_ratings = ratings.merge(movies,on='movieId',how='inner')
movie_ratings.head()
```

```python
# adding flag columns to see if a movie contains certain genres
for i in movie_genres:
    movie_ratings[i]=[1 if i in x else 0 for x in movie_ratings['genres']]
movie_ratings.head()
```

```python
# get columns needed to do a groupby for counting ratings by year for each genre
genre_cols = [x for x in movie_ratings.columns for y in movie_genres if y in x]
genre_and_year_cols = genre_cols + ['year_of_rating']

# create a filter to exclude ratings made in 1995, as there're only 3 ratings
exclude_year_1995 = movie_ratings['year_of_rating'] != 1995

# count ratings by year & genre
ratings_by_year = movie_ratings[exclude_year_1995][genre_and_year_cols].groupby('year_of_rating', as_index=True).sum()
print(ratings_by_year)
```

```python
# convert genre columns as rows to prepare for the visualization
# ratings_by_year=ratings_by_year.melt(['year_of_rating'], var_name='genres', value_name='ct_ratings')
# ratings_by_year.head()
```

```python
# visualize ratings of each genre by year
fig, ax = plt.subplots()
for genre in ratings_by_year.columns:
    ax.plot(ratings_by_year.index, ratings_by_year[genre], label=genre)
    plt.xlabel('Year of Ratings')
    plt.ylabel('Count of Ratings')
    plt.legend(loc='lower center',bbox_to_anchor=(1.3, -0.2),prop={'size': 9})
plt.show()
```
