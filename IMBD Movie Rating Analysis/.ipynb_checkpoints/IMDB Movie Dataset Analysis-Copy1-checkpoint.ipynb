### Load all the relevant files

```python
# read movies.csv file
movies = pd.read_csv('./ml-25m/movies.csv', sep=',')
print(type(movies))
movies.head()
```

```python
# read ratings.csv file
ratings = pd.read_csv('./ml-25m/ratings.csv', sep=',')
print(type(movies))
ratings.head()
```

### Data Prep

```python
# check if any of the timestamp is null
ratings['timestamp'].isnull().any()
```

```python
# convert unix time to datetime64
ratings['parsed_time'] = pd.to_datetime(ratings['timestamp'],unit='s')
ratings.head()
```

```python
#ratings['parsed_time'].min()
ratings['parsed_time'].max()
```

```python
# extract the year of ratings
ratings['year_of_rating'] = pd.DatetimeIndex(ratings['parsed_time']).year
ratings.head()
```

```python
# get the time range of the ratings - 1995 to 2019
ratings['year_of_rating'].describe()
```

```python
# check if any of the year has no enough rating records that cannot be counted as a good sample
ratings_count = ratings[['year_of_rating','rating']].groupby('year_of_rating').count()
print(ratings_count)
# below results indicate that only 3 ratings were made in 1995, so 1995 will be excluded from the analysis
```

```python
# see how the ratings are distributed
ratings['rating'].describe()
```

```python
# check if there's any null rating
ratings['rating'].isnull().any()
```

```python
# check if any null value in movie dataset
movies.isnull().any()
```

```python
# extract movie release year from title column in movies dataset
movies['year_of_release']=movies['title'].str.extract(r'\(([0-9]{4})\)')
movies.head()
```

```python
# split genres by using | as delimiter
movie_genres_df = movies['genres'].str.split('|',expand=True)
movie_genres_df.head(10)
```

```python
# split genres by using | as delimiter
# get all movie genres into a set (use union funcation to remove duplicates)
movie_genres = set()
for i in movies['genres'].to_list():
    genres_split = set(i.split('|'))
    movie_genres = movie_genres.union(genres_split)

# convert the set to list
movie_genres = list(movie_genres)
print(movie_genres)
```

```python
# remove perenthesis from movie_genres list
for i in movie_genres:
    movie_genres[movie_genres.index(i)] = i.replace('(','').replace(')','')
print(movie_genres)
```

### Data Analysis

```python
# create a filter to exclude ratings made in 1995 and 2019
# as there're only 3 ratings in 2015 and data is as of 2019 Nov 21st (no full data for 2019)
exclude_year_1995 = ratings['year_of_rating'] != 1995
exclude_year_2019 = ratings['year_of_rating'] != 2019

#create a filter to include only good movie ratings, which requires ratings higher or equal to 3.5
good_movies = ratings['rating'] >= 3.5
```

```python
# filter out 1995 and 2019 data, and only include good movie ratings
good_ratings = ratings[exclude_year_1995 & exclude_year_2019 & good_movies]
```

```python
# join movie dataset with rating dataset
movie_ratings = good_ratings.merge(movies,on='movieId',how='inner')
movie_ratings.head()
```

```python
# adding flag columns to see if a movie contains certain genres
for i in movie_genres:
    movie_ratings[i]=[1 if i in x else 0 for x in movie_ratings['genres']]
movie_ratings.head()
```

```python
# get relevant columns to do a groupby for counting ratings by year for each genre
genre_cols = [x for x in movie_ratings.columns for y in movie_genres if y in x]
genre_and_year_cols = genre_cols + ['year_of_rating']


# count ratings by year & genre
ratings_by_year = movie_ratings[genre_and_year_cols].groupby('year_of_rating', as_index=True).sum()
print(ratings_by_year)
```

```python
# convert genre columns as rows to prepare for the visualization
# ratings_by_year=ratings_by_year.melt(['year_of_rating'], var_name='genres', value_name='ct_ratings')
# ratings_by_year.head()
```

```python
# visualize ratings by year
fig, ax = plt.subplots(figsize=(20,10))
for genre in ratings_by_year.columns:
    ax.plot(ratings_by_year.index, ratings_by_year[genre], label=genre)
    plt.xlabel('Year of Ratings')
    plt.ylabel('Count of Ratings')
    plt.legend(loc='lower center',bbox_to_anchor=(1.1, 0.2),prop={'size': 14})
plt.show()
```

#### The above line plot has too many lines and it's hard to identify genres by colors.
#### So the following part will only select top 5 genres that received the most good ratings in each year to do visualization.

```python
# convert genre flag columns as rows to prepare for the sorting
ratings_by_year_2=ratings_by_year.reset_index().melt(['year_of_rating'], var_name='genres', value_name='ct_ratings')
ratings_by_year_2
```

```python
# sort genres by ct_ratings while pertaining asceding sorting of year_of_rating
# and only keep the top 5 genres for each year
ratings_sorted=ratings_by_year_2.set_index(['year_of_rating','genres']).groupby(level=0) \
               .apply(lambda x:x.groupby(level=1).sum().sort_values(by='ct_ratings',ascending=False).head())
ratings_sorted
```

```python
# convert top 5 genres back to columns
ratings_sorted = ratings_sorted.unstack(level=1)
ratings_sorted.head()
```

```python
# plot stacked bar chart
ratings_sorted.plot(kind='bar', stacked=True, figsize=(20,10))
```

```python
from matplotlib.pyplot import cm
import numpy as np

n = len(ratings_by_year.columns)  # count of unique genres
colors=cm.jet(np.linspace(0,1,n))  # generate colors that will be assigned to each genre in the next step


# visualize ratings by year for each genre
figure, axs = plt.subplots(n//4 if n%4==0 else n//4+1,4,figsize=(20,20))
for i,genre,color in zip(range(n),ratings_by_year.columns,colors):
    axs[i//4,i%4].plot(ratings_by_year.index, ratings_by_year[genre], label=genre,color=color)
    axs[i//4,i%4].set_title('Count of Ratings - '+genre, fontsize='15',fontweight='bold')
```

### Summaries and conclusions from the above analysis:
1. Romance ranked top 5 in 1996 and 1998-2002, it is then no longer among the top 5 since 2003.
This indicates that Romance movies are not as popular as it was before 2002.
2. Action, Comedy, Drama and Thriller are always in the top 5 list between 1996 and 2018. This means these 4 types remained as people's top favorites in the past 20 years.
3. Adventure ranked as top 5 in 1997 and after 2003. The number of good ratings Adventure movies received almost doubled from 1996 to 2018. This shows that Adventure gradually became a very popular type of movie since 2003. The number of good ratings doubled means the number of times this type of movie was seen increased a lot.
